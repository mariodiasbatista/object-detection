# Configure Object Detection Rotten Apple Dataset

```python
#Set the path vars
baseDir = '/content/scenario2_dataset/RottenApple_Project/dataset'
testDir = baseDir + '/test'
config = baseDir + '/../configs_model.yaml'

!ls {baseDir}
!ls {testDir}
!ls {config}
```

# Train Rotten Apple Dataset - First iteration

```python
!yolo task=detect mode=train data={config} epochs=20
```

# Evaluating the model - First iteration

```python
lastTrainPath = 'runs/detect/train'
!yolo task=detect mode=val model={lastTrainPath}/weights/best.pt data={config}
```
